```python
import pandas as pd
import ptitprince as pt  # raincloud
import seaborn as sns
import matplotlib.pyplot as plt

# Getting some data to play with is as easy as loading from seaborn
iris_df = sns.load_dataset("iris")

# Map categorical data
species_mapping = {'setosa':0,
    'versicolor':1,
    'virginica':2}
# creating a new column using map is sooo easy
iris_df['species_target'] = iris_df['species'].map(species_mapping)

# Making pretty default plots is easy in seaborn using styles
sns.set(style="white",font_scale=1.5)


dx="species"; dy="sepal_length"; ort="v"; pal = "Set2"; sigma = .2
ax=pt.RainCloud(x = dx, y = dy,
                data = iris_df,
                palette = 'Set2',
                width_viol = .4,
                width_box = .2,
                figsize = (7,5), orient = 'h',
               move = .0)
```

```python
import re
def preprocess(phrase):
    phrase = str(phrase) #convert to string
    phrase = phrase.lower() #lower case
    phrase = re.sub('[^a-zA-Z0-9 ]','',phrase) #remove special characters
    phrase = re.sub(' +',' ',phrase) #remove double spaces
    phrase = phrase.strip() #remove leading & trailing spaces
    return phrase

preprocess('         Apples   are fantastic!')
```

### Volcanic eruption on a map

```python
#!pip install folium
```

```python
from IPython.display import display
import folium
import branca.colormap as cm
import json
# create map centred on eruption source location
lat, lon = (-4.079999923706055,145.0399932861328)
m = folium.Map(location=[lat, lon], zoom_start=7)
# add eruption source location marker
folium.Marker([lat, lon], tooltip='Source location').add_to(m)
# add eruption load contours
with open('eruption.json') as fp:
    geojson = json.load(fp)
# folium geojson style function to pull out imperfect match on style properties
style_function = lambda x: { 'color': x['properties']['stroke'], 'fillColor': x['properties']['fill'], 'opacity': x['properties']['stroke-opacity'], 'fillOpacity': x['properties']['fill-opacity'] }
folium.GeoJson(geojson, name='geojson', style_function=style_function).add_to(m)
# add legend
geojson_dict = json.loads(geojson)
colours_in_bar = [x['properties']['stroke'] for x in geojson_dict['features']]
step_index = list(range(5,105,5))
step = cm.StepColormap(colours_in_bar, vmin=0, vmax=100, index=step_index, caption='%')
m.add_child(step)
display(m)
```
